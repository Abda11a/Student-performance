## Inferential staticts

```python
# Ckeck the probability of students who score more than 50 marsk in each


total_students = data.shape[0]
math_more_than_50= data[data['math score']>50].shape[0]
readind_more_than_50= data[data['reading score']>50].shape[0]
writting_more_than_50= data[data['writing score']>50].shape[0]


math_prb_more_50 = (math_more_than_50/total_students)*100
reading_prb_more_50 = (readind_more_than_50/total_students)*100
writing_prb_more_50 = (writting_more_than_50/total_students)*100


print(f"Probability of Students Scoring more than 50 marks in *Maths* : {math_prb_more_50}")
print(f"Probability of Students Scoring more than 50 marks in *reading* : {reading_prb_more_50}")
print(f"Probability of Students Scoring more than 50 marks in *writing* : {writing_prb_more_50}")
```

#### Lets check the Inference

```python
np.random.seed(6)

sample_math = np.random.choice(a=data['math score'], size=100)

print (f"Sample math mean = {sample_math.mean()}")
print (f"Population Math mean = {data['math score'].mean()}")


sample_writing = np.random.choice(a=data['writing score'], size=100)
print(f"Sample writting mean =  {sample_writing.mean()}")
print(f"Population writting mean = {data['writing score'].mean()}")
```

##### check the Confidence Interval for Reading Score

```python
import scipy.stats as stats
import math

# lets seed the random values
np.random.seed(10)

# lets take a sample size
sample_size = 1000
sample = np.random.choice(a= data['reading score'],
                          size = sample_size)
sample_mean = sample.mean()

# Get the z-critical value*
z_critical = stats.norm.ppf(q = 0.95)  

 # Check the z-critical value  
print("z-critical value: ",z_critical)                                

# Get the population standard deviation
pop_stdev = data['reading score'].std()  

# checking the margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

# defining our confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

# lets print the results
print("Confidence interval:",end=" ")
print(confidence_interval)
print("True mean: {}".format(data['reading score'].mean()))
```
